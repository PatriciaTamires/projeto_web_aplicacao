## Descrição do Projeto

A proposta deste projeto é desenvolver um aplicativo de entrega de comida completo e escalável. A plataforma permitirá que restaurantes se cadastrem e gerenciem seus cardápios, enquanto os clientes poderão navegar pelas opções disponíveis, realizar pedidos e acompanhar o status da entrega em tempo real.

O sistema contará com diversas funcionalidades essenciais, incluindo:

Cadastro e gerenciamento de restaurantes

Cadastro e organização de produtos (comidas e bebidas)

Serviço de pedidos e entregas

Sistema de avaliações e feedback dos usuários

Acompanhamento do status do pedido

O objetivo é criar uma solução intuitiva, eficiente e de fácil utilização tanto para clientes quanto para estabelecimentos parceiros.

```python
import os

##Cadastro restaurantes

def exibir_opcoes():
    print('1. Cadastro restaurantes')
    print('2. Listar restaurantes')
    print('3. Ativar restaurante')
    print('4. Sair')

def sair():
    os.system('cls')
    print('Finalizando App\n')

def escolher_opcoes():
    opcao_escolhida = input('Escolha uma opção')
    print('Você escolheceu a opção', opcao_escolhida)

    if opcao_escolhida == '1':
        print('Cadastro de restaurantes')
    elif opcao_escolhida == '2':
        print('Listagem de restaurantes')
    elif opcao_escolhida == '3':
        print('Ativar restaurantes')
    else:
        sair()

def main():
    exibir_opcoes()
    escolher_opcoes()

if __name__ == '__main__':
    main()
```
